```python
import matplotlib.pyplot as plt
x1 = [0, 4, 5, 1, 0]
y1 = [0, 0, 3, 3, 0]

plt.plot(x1, y1, color="green", linestyle="-.", marker="*", markersize=10)

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Parallelogram ABCD')

plt.xlim(-1, 6)
plt.ylim(-1, 4)
plt.show()
```

```python
x2 = [1, 5, 3, 1]
y2 = [1, 1, 4, 1]
plt.plot(x2, y2, color="green", linestyle="--", marker="o", markersize=10, markerfacecolor="red")
plt.title('Isosceles Triangle')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.xlim(0, 6)
plt.ylim(0, 5)
plt.show()
```

```python
x = [3, 2, 1.8, 2, 3, 3.2,3,2]
y = [4, 4, 3.5, 3, 3, 2.5,2,2]
plt.plot(x, y, linewidth=5,marker="*")
plt.xlim(0, 5)
plt.ylim(0, 5)
plt.title("Plot of the Letter 'S'")
plt.show()
```

```python
students = ['Ava', 'Alex', 'Jules', 'Josh', 'Rhys']
credits = [120, 95, 110, 135, 105]

bar = ['skyblue', 'lightgreen', 'yellow', 'cyan', 'orange']
plt.bar(students, credits, color=bar, width=0.7)
plt.xlabel('Students', color='blue')
plt.ylabel('Total Credits Earned', color='green')
plt.title('Total Credits Earned by Students', color='purple')
plt.show()
```
