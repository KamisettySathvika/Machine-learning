```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('port_data.csv')

# Get the top 10 countries by vessels in port
top_countries = df.nlargest(10, 'Vessels in Port')


plt.figure(figsize=(18, 12))
plt.suptitle('Port Data Insights')

# Plot 1: Vessels in Port per Country
plt.subplot(2, 3, 1)
plt.bar(top_countries['Country'], top_countries['Vessels in Port'], color='skyblue')
plt.title('Vessels in Port per Country')
plt.xlabel('Country')
plt.ylabel('Vessels in Port')
plt.xticks(rotation=45)  

# Plot 2: Departures (Last 24 Hours) per Country
plt.subplot(2, 3, 2)
plt.bar(top_countries['Country'], top_countries['Departures(Last 24 Hours)'], color='lightgreen')
plt.title('Departures (Last 24 Hours) per Country')
plt.xlabel('Country')
plt.ylabel('Departures (Last 24 Hours)')
plt.xticks(rotation=45) 

# Plot 3:Distribution of Vessels in Port by Country
plt.subplot(2, 3, 3)
country_counts = df['Country'].value_counts().head(10)
plt.pie(country_counts, labels=country_counts.index)
plt.title('Distribution of Vessels in Port by Country')

# Plot 4: Expected Arrivals per Country
plt.subplot(2, 3, 4)
plt.bar(top_countries['Country'], top_countries['Expected Arrivals'], color='gold')
plt.title('Expected Arrivals per Country')
plt.xlabel('Country')
plt.ylabel('Expected Arrivals')
plt.xticks(rotation=45) 

# Plot 5: Departures vs Arrivals (Last 24 Hours)
plt.subplot(2, 3, 5)
plt.scatter(top_countries['Departures(Last 24 Hours)'], top_countries['Arrivals(Last 24 Hours)'], color='purple')
plt.title('Departures vs Arrivals (Last 24 Hours)')
plt.xlabel('Departures (Last 24 Hours)')
plt.ylabel('Arrivals (Last 24 Hours)')


# Plot 6: Vessels in Port vs Expected Arrivals
plt.subplot(2, 3, 6)
plt.scatter(top_countries['Vessels in Port'], top_countries['Expected Arrivals'], color='orange')
plt.title('Vessels in Port vs Expected Arrivals')
plt.xlabel('Vessels in Port')
plt.ylabel('Expected Arrivals')
  

plt.tight_layout()
plt.show()
```
