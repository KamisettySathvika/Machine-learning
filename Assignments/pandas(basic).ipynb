```python
import pandas as pd
```

```python
df=pd.read_csv("Port_Data.csv")

df.head(3)
```

```python
df.shape
```

```python
#if the country name is china and the type is anchorage

df['Country'] = df['Country'].str.strip().str.capitalize()
df['Type'] = df['Type'].str.strip().str.capitalize()

f1 = df[(df['Country'] == 'China') & (df['Type'] == 'Anchorage')]
print(f1.head(3))
```

```python
# if arrivals are greater than the departures by a 100
f2=df[df["Arrivals(Last 24 Hours)"]-df["Departures(Last 24 Hours)"]>100]
f2.head(3)
```

```python
#create an array where null value rows are deleted
f3=df.dropna()
print(f3.head())

#create an array where null value cols are deleted
f4=df.dropna(axis=1)
print(f4.head(3))
```

```python
#if the vessles in port and greater than a 1000 or the depatures are greater than 200
f5=df[(df["Vessels in Port"] > 1000) | (df["Departures(Last 24 Hours)"] > 200)]
f5.head()
```

```python
#where expected arrivals are > actual arrivals print middle ten rows with just 3 cols

import pandas as pd
f6 = df[df["Expected Arrivals"] > df["Arrivals(Last 24 Hours)"]]

tr = len(f6)
midd = tr // 2
start= midd- 5  
end= midd + 5  

middle= f6.iloc[start:end + 1, 0:3]  
print(middle)
```

```python
#print out the median / mode value for Expected arrivals
f7=df["Expected Arrivals"].mean()
f8=df["Expected Arrivals"].median()
f9=df["Expected Arrivals"].mode()
print("Mean=",f7)
print("Median=",f8)
print("Mode=",f9)
```

```python
# print out the top 5 countries with the highest gap between the departures and arrivals
df['gap'] = df['Departures(Last 24 Hours)'] - df['Arrivals(Last 24 Hours)']
top= df.sort_values(by='gap', ascending=False).head(5)
f10 = top[['Country', 'Departures(Last 24 Hours)', 'Arrivals(Last 24 Hours)', 'gap']]
print(f10)
```

```python
#print out the top 3 ports by number of departures for only the countries that have more than 5 ports
for name, c_df in df.groupby("Country"):
    if len(c_df) > 5:
        print("\nCountry:", name)
        top_ports = c_df["Departures(Last 24 Hours)"].nlargest(3).index
        for index in top_ports:
            print(c_df.at[index, "Port Name"])
```
