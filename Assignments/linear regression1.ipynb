```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv("./weather.csv")[:100]
```

```python
df["Humidity"].head(),df["Temperature (C)"].head()
```

```python
plt.scatter(df["Temperature (C)"], df["Humidity"], label='Actual Humidity')
plt.ylabel("Humidity")
plt.xlabel("Temperature (C)")

weight = 1
bias = -1

def mse(y, y_):
    return sum((y - y_) ** 2) / len(y)


def grad(y, y_, x):
    return sum(-2 * (y - y_) * x) / len(y), sum(-2 * (y - y_)) / len(y)

def pred(w, b, x):
    return w * x + b

lr = 0.005

for i in range(1000):
    feature = df["Temperature (C)"]
    label = df["Humidity"]
    
    prediction = pred(weight, bias, feature)
    loss = mse(label, prediction)
    w_grad, b_grad = grad(label, prediction, feature)
    
    weight -= w_grad * lr
    bias -= b_grad * lr

print(f"Loss = {loss:.4f}, Weight = {weight:.4f}, Bias = {bias:.4f}")


df["Prediction"] = pred(weight, bias, df["Temperature (C)"])
plt.plot(df["Temperature (C)"], df["Prediction"], c="red")
plt.show()

temp = 8.288889
predicted_humidity = pred(weight, bias, temp)
print(f"Predicted humidity for temperature {temp}°C is {predicted_humidity:.4f}")
```
